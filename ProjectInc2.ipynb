```python
def inflation_rate(low_rate,high_rate):
  filtered_countries = []
  filtered_inflation_rates = []
  count=0
  for i in range(len(data.country)):
     if data.inflation[i] <high_rate and data.inflation[i] >low_rate:
        filtered_countries.append(data.country[i])
        filtered_inflation_rates.append(data.inflation[i])
        count=count+1
   # create a bar chart with the filtered data
  plt.bar(filtered_countries, filtered_inflation_rates, linewidth=9,align='edge',width=0.5) 
  # add labels and title
  plt.xlabel("Countries")
  plt.ylabel("Inflation Rate")
  str
  plt.title("Inflation Rates by Country (less than "+str(high_rate) +") and greater than "+ str(low_rate) +" is "   + str(count))

  plt.xticks(rotation=90,fontsize=8)

  # show the plot
  plt.show()
```
